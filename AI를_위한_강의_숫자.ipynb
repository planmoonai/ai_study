float64 vs float32 정밀도 비교

--------------------
float32 의 정밀도: 유효숫자 자리수 7자리

float64 의 정밀도: 유효숫자 자리수 15~16자리


```python
# 64비트 부동소수점 (double precision)
a = np.float64(1.00000001)
b = np.float64(1.00000000)
print(a - b)
# 출력: 9.99999993922529e-09

# 32비트 부동소수점 (single precision)
a = np.float32(1.00000001)
b = np.float32(1.00000000)
print(a - b)
# 출력: 0.0
```

```python
sum = 10000000.0   #매우 큰숫자
for i in range(10000000):
  sum += 0.0000000001   #매우작은 숫자를 for함수에서 계속 더하기

print(sum) #100000000.0

#실수하기 좋음. 아주큰숫자에 아주작은숫자를 더하면 누적이 안되는경우가 있다.
```

```python
# 오류피하기 : 처음에 0으로 시작하기!

sum = 0.0   #0으로 시작
for i in range(10000000):
  sum += 0.0000000001          #작은숫자를 매우여러번 더하기
sum += 10000000.0  #그다음에 큰숫자 더하기

print(sum)  #10000000.001 작은숫자부분이 무시되지않음.

#뉴럴네트워크를 훈련시킬때 이런 정밀도 때문에 골치 아팠던적도 있었음. 요새는 해결할수있는 다양한 기술이 발전함.
```
